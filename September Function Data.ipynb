```python
def turnstiledata_clean(df):

    cols = []
    for column in df.columns:
        cols.append(column.lower().replace(" ",''))
    df.columns = cols
    
    df['entries_hours'] = df['entries'].shift(1)
    df['entries_per_4hour'] = df['entries']-df['entries_hours']
    
    df['exits_hours'] = df['exits'].shift(1)
    df['exits_per_4hour'] = df['exits']-df['exits_hours']
   
    df.drop(["entries", 'exits', 'entries_hours', 'desc', 'exits_hours'], axis=1, inplace=True)
    
    df = df.fillna(0)

    df['date'] = pd.to_datetime(df['date'])
    df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.time

    return df
```

```python
clean_data=turnstiledata_clean(turnstile_data)
```

```python
os.listdir('./MTA Weekly Data 2019')
```

```python
complete_year = clean_data    
for file in os.listdir('./MTA Weekly Data 2019'):
    if file.endswith(".txt"):
        print(file)
        data = pd.read_csv('./MTA Weekly Data 2019/'+ file)
        data = turnstiledata_clean(data)
        print(data.shape)
        complete_year=pd.concat([complete_year,data], axis=0, ignore_index=True)
```
